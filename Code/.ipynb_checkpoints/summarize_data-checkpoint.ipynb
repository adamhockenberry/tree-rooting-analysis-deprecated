```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import matplotlib
matplotlib.rcParams['xtick.labelsize'] = 12
matplotlib.rcParams['ytick.labelsize'] = 12
matplotlib.rcParams['axes.labelsize'] = 14
matplotlib.rcParams['axes.titlesize'] = 14

matplotlib.rcParams['axes.grid'] = True
matplotlib.rcParams['grid.color'] = '0.5'
matplotlib.rcParams['grid.linewidth'] = '0.5'

matplotlib.rcParams['axes.edgecolor'] = '0.25'
matplotlib.rcParams['xtick.color'] = '0'
matplotlib.rcParams['ytick.color'] = '0'

matplotlib.rcParams['xtick.major.width'] = 1
matplotlib.rcParams['ytick.major.width'] = 1
matplotlib.rcParams['ytick.major.size'] = 5
matplotlib.rcParams['xtick.major.size'] = 5
matplotlib.rcParams['axes.spines.right'] = True
matplotlib.rcParams['axes.spines.left'] = True
matplotlib.rcParams['axes.spines.top'] = True
matplotlib.rcParams['axes.spines.bottom'] = True

matplotlib.rcParams['font.family'] = 'sans-serif'
matplotlib.rcParams['font.sans-serif'] = 'Helvetica'
matplotlib.rcParams['font.weight']='normal'
matplotlib.rcParams['axes.axisbelow'] = True
```

```python
monophyly_df = pd.read_csv('../Data/OMA_group_data/eukaryotes/tree_monophyly.csv', index_col='Group_id')
monophyly_df_nan = monophyly_df[monophyly_df['monophyletic_clade'].isnull()==False]
# monophyly_df = pd.read_csv('../Data/Tria_et_al_data/eukaryotes/tree_monophyly.csv', index_col='Group_id')
print(monophyly_df.shape)
monophyly_df.head(n=10)
```

```python
fig, ax = plt.subplots()
ax.hist(monophyly_df_nan['monophyletic_n']+monophyly_df_nan['other_n']);
ax.set_xlabel('Total number of species')
```

```python
fig, ax = plt.subplots()
ax.hist(monophyly_df_nan['monophyletic_total_bl']+monophyly_df_nan['other_total_bl']);
ax.set_xlabel('Total branch length')
```

```python
fig, ax = plt.subplots()
ax.hist((monophyly_df_nan['monophyletic_total_bl']+monophyly_df_nan['other_total_bl'])/\
        (monophyly_df_nan['monophyletic_n']+monophyly_df_nan['other_n']));
ax.set_xlabel('Average branch length per species')
```

```python
fig, ax = plt.subplots()
ax.plot(monophyly_df_nan['monophyletic_n'], monophyly_df_nan['other_n'], 'bo')
ax.plot([40, 145], [40, 145], 'k-')
ax.set_xlabel('Number of species ()')
ax.set_ylabel('Number of species ()')
```

```python
fig, ax = plt.subplots()
ax.plot(monophyly_df_nan['monophyletic_total_bl'], monophyly_df_nan['other_total_bl'], 'bo')
ax.plot([0, 80], [0, 80], 'k-')

ax.set_xlabel('Total branch length ()')
ax.set_ylabel('Total branch length ()')
```

```python
fig, ax = plt.subplots()
ax.plot(monophyly_df_nan['monophyletic_total_bl']/monophyly_df_nan['monophyletic_n'],\
        monophyly_df_nan['other_total_bl']/monophyly_df_nan['other_n'], 'bo')
ax.plot([0, 1], [0, 1], 'k-')

ax.set_xlabel('Average branch length per species ()')
ax.set_ylabel('Average branch length per species ()')
```
